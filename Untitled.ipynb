```python
import random
import json
import os
from pathlib import Path
from typing import Union, List

def load_all_jsonl_files(input_source: Union[str, Path, List[str], List[Path]]):
    """Yields JSON objects from all .jsonl files in the input path(s)."""
    if isinstance(input_source, (str, Path)):
        input_source = Path(input_source)
        if input_source.is_dir():
            jsonl_files = sorted(input_source.glob("*.jsonl"))
        elif input_source.is_file():
            jsonl_files = [input_source]
        else:
            raise ValueError(f"Input path {input_source} is neither a directory nor a valid file.")
    elif isinstance(input_source, list):
        jsonl_files = [Path(p) for p in input_source]
    else:
        raise TypeError("Input must be a path, file, or list of file paths.")

    for filepath in jsonl_files:
        with open(filepath, 'r') as f:
            for line in f:
                yield json.loads(line)

def save_chunk(chunk_data, output_dir, chunk_idx):
    """Save a list of packed examples to a chunked output file."""
    os.makedirs(output_dir, exist_ok=True)
    out_path = Path(output_dir) / f"packed_{chunk_idx:04d}.jsonl"
    with open(out_path, 'w') as f:
        for item in chunk_data:
            f.write(json.dumps(item) + '\n')
    print(f"✅ Saved {len(chunk_data)} packed examples to {out_path}")

def pack_examples_from_input(input_source,
                             output_dir,
                             max_seq_length=2048,
                             eos_token_id=None,
                             chunk_size=3000,
                             start_pack_idx=0,
                             shuffle=True) -> int:
    """
    Pack tokenized examples into blocks with max token length.

    Returns:
        next_pack_idx: next available unique pack ID
    """
    dataset = load_all_jsonl_files(input_source)
    dataset_list = list(dataset)
    print(f"📦 Found {len(dataset_list)} examples to process.")
    if shuffle:
        random.shuffle(dataset_list)

    packed = []
    chunk_counter = 0
    current_pack_idx = start_pack_idx

    current_pack = {
        "tokens": [],
        "example_ids": [],
        "rule_ids": [],
        "packed_seq_lengths": [],
        "length": 0
    }
    current_len = 0

    # Reserve space for EOS token if needed
    max_pack_len = max_seq_length - 1 if eos_token_id is not None else max_seq_length

    for item in dataset_list:
        tokens = item["tokens"]
        length = item["length"]  # Use provided length

        if length > max_pack_len:
            print(f"⚠️ Skipping example id={item['id']} (length {length} > max {max_pack_len})")
            continue

        if current_len + length <= max_pack_len:
            current_pack["tokens"].extend(tokens)
            current_pack["example_ids"].append(item["id"])
            current_pack["rule_ids"].append(item["rule_id"])
            current_pack["packed_seq_lengths"].append(length)
            current_len += length
            current_pack["length"] = current_len
        else:
            # Finalize current pack
            if eos_token_id is not None:
                current_pack["tokens"].append(eos_token_id)
                current_pack["length"] += 1
                current_len += 1
                current_pack["packed_seq_lengths"][-1] += 1  # Update length of last example

            current_pack["pack_id"] = current_pack_idx
            packed.append(current_pack)
            current_pack_idx += 1

            if len(packed) >= chunk_size:
                save_chunk(packed, output_dir, chunk_counter)
                packed = []
                chunk_counter += 1

            # Start new pack
            current_pack = {
                "tokens": tokens[:],
                "example_ids": [item["id"]],
                "rule_ids": [item["rule_id"]],
                "packed_seq_lengths": [length],
                "length": length
            }
            current_len = length

    # Final pack
    if current_pack["tokens"]:
        if eos_token_id is not None:
            current_pack["tokens"].append(eos_token_id)
            current_pack["length"] += 1
            current_pack["packed_seq_lengths"][-1] += 1
        current_pack["pack_id"] = current_pack_idx
        packed.append(current_pack)
        current_pack_idx += 1

    if packed:
        save_chunk(packed, output_dir, chunk_counter)

    return current_pack_idx  # Return the next available unique pack ID

```

```python
import os
import json
from pathlib import Path
from typing import Union, List

def count_tokens_and_unique_rule_ids(input_path_or_files: Union[str, List[str]],
                                     token_key="tokens",
                                     rule_id_key="rule_id"):
    """
    Count total tokens and collect unique rule_ids from a directory or list of .jsonl files.

    Parameters:
        input_path_or_files (str | List[str]): Directory path or list of file paths.
        token_key (str): Key for tokens list in JSON objects.
        rule_id_key (str): Key for rule_id in JSON objects.

    Returns:
        total_tokens (int): Total number of tokens.
        total_examples (int): Number of examples processed.
        unique_rule_ids (set): Set of unique rule_ids found.
    """
    if isinstance(input_path_or_files, str):
        file_list = sorted(Path(input_path_or_files).glob("*.jsonl"))
    else:
        file_list = [Path(p) for p in input_path_or_files]

    total_tokens = 0
    total_examples = 0
    unique_rule_ids = set()

    for file_path in file_list:
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f:
                try:
                    obj = json.loads(line)
                    tokens = obj.get(token_key, [])
                    rule_id = obj.get(rule_id_key)
                    total_tokens += len(tokens)
                    total_examples += 1
                    if rule_id is not None:
                        unique_rule_ids.add(rule_id)
                except Exception as e:
                    print(f"⚠️ Skipping line in {file_path} due to error: {e}")

    print(f"📄 Total examples: {total_examples}")
    print(f"🧮 Total tokens: {total_tokens}")
    print(f"🔑 Unique rule_ids: {len(unique_rule_ids)}")

    return total_tokens, total_examples, unique_rule_ids

```

```python
from collections import defaultdict
from tqdm import tqdm
import os
import random
import orjson

def load_all_jsonl_files(file_path):
    with open(file_path, 'rb') as f:
        for line in f:
            try:
                yield orjson.loads(line)
            except orjson.JSONDecodeError:
                continue

def chunked(iterable, n):
    """Yield successive n-sized chunks from iterable."""
    for i in range(0, len(iterable), n):
        yield iterable[i:i + n]

def sample_by_rule_id(input_source, sample_per_rule=5, batch_size=10):
    """
    Sample up to N examples per rule_id from each file, streaming 10 files at a time.

    Returns:
        sampled_data: list of sampled items
        seen_rule_ids: set of rule_ids used
    """
    if isinstance(input_source, str) and os.path.isdir(input_source):
        files = [os.path.join(input_source, f) for f in os.listdir(input_source) if f.endswith('.jsonl')]
    elif isinstance(input_source, list):
        files = input_source
    else:
        raise ValueError("input_source should be a directory path or a list of file paths.")

    seen_rule_ids = set()
    all_samples = []

    for file_batch in tqdm(list(chunked(files, batch_size)), desc="📦 Processing file batches"):
        for file in file_batch:
            rule_to_buffer = defaultdict(list)

            with open(file, 'rb') as f:
                for line in f:
                    try:
                        item = orjson.loads(line)
                    except orjson.JSONDecodeError:
                        continue

                    rule_id = item.get("rule_id")
                    if rule_id is None:
                        continue

                    if len(rule_to_buffer[rule_id]) < sample_per_rule:
                        rule_to_buffer[rule_id].append(item)

            for rule_id, examples in rule_to_buffer.items():
                all_samples.extend(examples)
                seen_rule_ids.add(rule_id)

            rule_to_buffer.clear()

    return all_samples, seen_rule_ids



def pack_examples_from_input_with_sampling(input_source,
                                           output_dir,
                                           max_seq_length=2048,
                                           eos_token_id=None,
                                           chunk_size=3000,
                                           start_pack_idx=0,
                                           sample_per_rule=5,
                                           shuffle=True):
    """
    Sample examples grouped by rule_id (up to 5 per rule per file), then pack them.

    Returns:
        next_pack_idx: next available unique pack ID
        seen_rule_ids: set of all rule_ids used
    """
    dataset_list, seen_rule_ids = sample_by_rule_id(input_source, sample_per_rule=sample_per_rule)
    print(f"📦 Sampled {len(dataset_list)} examples from rule_ids: {len(seen_rule_ids)} unique rules")

    if shuffle:
        random.shuffle(dataset_list)

    packed = []
    chunk_counter = 0
    current_pack_idx = start_pack_idx

    current_pack = {
        "tokens": [],
        "example_ids": [],
        "rule_ids": [],
        "packed_seq_lengths": [],
        "length": 0
    }
    current_len = 0
    max_pack_len = max_seq_length - 1 if eos_token_id is not None else max_seq_length

    for item in dataset_list:
        tokens = item["tokens"]
        length = item["length"]

        if length > max_pack_len:
            print(f"⚠️ Skipping example id={item['id']} (length {length} > max {max_pack_len})")
            continue

        if current_len + length <= max_pack_len:
            current_pack["tokens"].extend(tokens)
            current_pack["example_ids"].append(item["id"])
            current_pack["rule_ids"].append(item["rule_id"])
            current_pack["packed_seq_lengths"].append(length)
            current_len += length
            current_pack["length"] = current_len
        else:
            if eos_token_id is not None:
                current_pack["tokens"].append(eos_token_id)
                current_pack["length"] += 1
                current_len += 1
                current_pack["packed_seq_lengths"][-1] += 1

            current_pack["pack_id"] = current_pack_idx
            packed.append(current_pack)
            current_pack_idx += 1

            if len(packed) >= chunk_size:
                save_chunk(packed, output_dir, chunk_counter)
                packed = []
                chunk_counter += 1

            current_pack = {
                "tokens": tokens[:],
                "example_ids": [item["id"]],
                "rule_ids": [item["rule_id"]],
                "packed_seq_lengths": [length],
                "length": length
            }
            current_len = length

    if current_pack["tokens"]:
        if eos_token_id is not None:
            current_pack["tokens"].append(eos_token_id)
            current_pack["length"] += 1
            current_pack["packed_seq_lengths"][-1] += 1
        current_pack["pack_id"] = current_pack_idx
        packed.append(current_pack)
        current_pack_idx += 1

    if packed:
        save_chunk(packed, output_dir, chunk_counter)

    return current_pack_idx, seen_rule_ids
```

```python
import json
import os
from pathlib import Path


def load_jsonl_file(filepath):
    """Yields JSON objects from a single .jsonl file."""
    with open(filepath, 'r') as f:
        for line in f:
            yield json.loads(line)

def save_chunk(chunk_data, output_dir, chunk_idx):
    """Save a list of packed examples to a chunked output file."""
    os.makedirs(output_dir, exist_ok=True)
    out_path = Path(output_dir) / f"packed_{chunk_idx:04d}.jsonl"
    with open(out_path, 'w') as f:
        for item in chunk_data:
            f.write(json.dumps(item) + '\n')
    print(f"✅ Saved {len(chunk_data)} packed examples to {out_path}")

def pack_examples_from_dir(input_dir, output_dir, max_seq_length=2048, eos_token_id=None, chunk_size=10000, start_id = 0):
    jsonl_files = sorted(Path(input_dir).glob("*.jsonl"))
    i = start_id

    packed = []
    current_pack = {
        "id": i,
        "tokens": [],
        "example_ids": [],
        "rule_ids": [],
        "length": 0
    }
    current_len = 0
    chunk_counter = 0

    # Reserve one token slot for EOS if provided
    max_pack_len = max_seq_length - 1 if eos_token_id is not None else max_seq_length

    for filepath in jsonl_files:
        for item in load_jsonl_file(filepath):
            tokens = item["tokens"]
            length = len(tokens)

            if length > max_pack_len:
                print(f"⚠️ Skipping example id={item['id']} (length {length} > max {max_pack_len})")
                continue

            # Check if it fits in current pack
            if current_len + length <= max_pack_len:
                current_pack["tokens"].extend(tokens)
                current_pack["example_ids"].append(item["id"])
                current_pack["rule_ids"].append(item["rule_id"])
                current_len += length
                current_pack["length"] = current_len
            else:
                # Finalize and save current pack
                if eos_token_id is not None:
                    current_pack["tokens"].append(eos_token_id)
                    current_pack["length"] += 1

                packed.append(current_pack)
                if len(packed) >= chunk_size:
                    save_chunk(packed, output_dir, chunk_counter)
                    packed = []
                    chunk_counter += 1

                # Start new pack
                i += 1
                current_pack = {
                    "id": i,
                    "tokens": tokens[:],
                    "example_ids": [item["id"]],
                    "rule_ids": [item["rule_id"]],
                    "length": length
                }
                current_len = length

    # Final pack (with EOS token if provided)
    if current_pack["tokens"]:
        if eos_token_id is not None:
            current_pack["tokens"].append(eos_token_id)
            current_pack["length"] += 1
        packed.append(current_pack)

    # Save any remaining packed data
    if packed:
        save_chunk(packed, output_dir, chunk_counter)
    return i


# ==== Example usage ====
if __name__ == "__main__":
    pack_examples_from_dir(
        input_dir="data/input_jsonls",
        output_dir="data/packed_output",
        max_seq_length=2048,
        eos_token_id=50256,
        chunk_size=1000
    )
```

```python
import os
import json

def reindex_jsonl_directory(directory_path, id_key="id", id_template="dev_{index}"):
    index = 0
    for filename in sorted(os.listdir(directory_path)):
        if filename.endswith(".jsonl"):
            file_path = os.path.join(directory_path, filename)
            new_lines = []

            with open(file_path, "r", encoding="utf-8") as infile:
                for line in infile:
                    data = json.loads(line)
                    data[id_key] = id_template.format(index=index)
                    new_lines.append(json.dumps(data))
                    index += 1

            with open(file_path, "w", encoding="utf-8") as outfile:
                for new_line in new_lines:
                    outfile.write(new_line + "\n")

    print(f"Re-indexed {index} lines using template '{id_template}' in '{directory_path}'.")

```
