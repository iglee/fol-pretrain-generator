```python
def write_list_to_file(my_list, file_name):
    """
    Writes each item in a list to a file, each on a new line.
    
    :param my_list: List of items to be written into the file.
    :param file_name: The name of the file to write the list to.
    """
    with open(file_name, 'w') as file:
        for item in my_list:
            file.write(str(item) + '\n')

    print(f"List has been written to {file_name}")

# Example usage
my_list = ["apple", "banana", "cherry", "date"]
file_name = "my_list.txt"

write_list_to_file(my_list, file_name)


import re

def remove_duplicate_predicates(predicates):
    seen = set()
    result = []

    for pred in predicates:
        match = re.match(r'^(\w+)\((.*?)\)$', pred)
        if match:
            func_name = match.group(1)
            args = match.group(2).split(',')
            key = (func_name, len(args))
            if key not in seen:
                seen.add(key)
                result.append(pred)
    return result

```
