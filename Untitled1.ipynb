```python
import ijson
import matplotlib.pyplot as plt
from collections import Counter
from src.fol_util import read_pkl_fie, standardize_variables, test_train_split_list, write_list_to_json


counter = Counter()

with open('fol-rules/train_rules.json', 'r') as f:
    for obj in ijson.items(f, 'item'):
        complexity_list = obj.get('complexity_by_step', [])
        if complexity_list:  # make sure it's not empty
            first_value = complexity_list[0]
            counter[first_value] += 1

# Now plot
plt.bar(counter.keys(), counter.values())
plt.xlabel('First Complexity Value')
plt.ylabel('Frequency')
plt.title('Histogram of First Complexity Values')
plt.show()
```

```python
import ijson
import random
from collections import defaultdict

# Dictionary to store data points grouped by their first complexity value
complexity_groups = defaultdict(list)

# Open the file and stream it
with open('fol-rules/train_rules.json', 'r') as f:
    for obj in ijson.items(f, 'item'):
        complexity_list = obj.get('complexity_by_step', [])
        if complexity_list:  # Ensure it's not empty
            first_value = complexity_list[0]
            complexity_groups[first_value].append(obj)

# Randomly pick 30 data points for each complexity value
random_samples = {}
for complexity_value, data_points in complexity_groups.items():
    if len(data_points) >= 30:
        random_samples[complexity_value] = random.sample(data_points, 30)
    else:
        random_samples[complexity_value] = data_points  # If less than 30, take all

# Print the sampled data points for each complexity value
for complexity_value, samples in random_samples.items():
    print(f"Complexity Value: {complexity_value}")
    print(f"Sampled Data Points: {samples[:5]}...")  # Print first 5 for brevity
    print("-" * 40)
```
